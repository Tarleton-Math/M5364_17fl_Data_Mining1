Now, we meet our first machine learning technique: $k$-nearest neighbors.  The idea is fairly simple and easiest to describe when all the features (aka independent/input/explanatory/predictor variables) are quantitative and the target (aka dependent/output/class/label) variable is categorical. 

Assume we have $n$ observations of $m$ quantitative features and 1 categorical target variable.  We want to build a gadget (called a classifier) that will help us predict the class of future observations where we know the $m$ features, but NOT the class.

Steps:
1. Plot each observation (row) as a point in $m$ dimensional space.  Thus, you have $n$ points living in $\mathbb{R}^m$.
1. Label/color each point by its class.
1. You now receive a new observation with the $m$ features, but not the class.  Plot it as another point in $\mathbb{R}^m$.
1. Find the $k$ points from the old data that are closest to the new point.
1. The most frequent class among these points is the predicted class for the new observation.
  1. We can give a more refined result called the posterior distribution.  We can say "30% of the $k$ nearest neighbors voted for class 0, 20% voted for class 1, 50% voted for class 2."  Clearly class 2 wins, but we have quantified our confidence in this classification - about 50-50.  The posterior distribuion is [0.3, 0.2, 0.5].
